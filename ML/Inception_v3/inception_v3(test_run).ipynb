그래프 생성

```
create_graph()
predict = 0
prediction_accuracy = 0
cat_dog_path = 'cats'
```

```
# total_test_num = len(x_test)
total_test_num = 100
for test_num in range(total_test_num):
  if x_test[test_num].split('.')[0] == 'cat' :
    image_data = tf.gfile.FastGFile(test_path + '/' +image_path[0] + '/' + x_test[test_num], 'rb').read()
    cat_dog_path = 'cats'
    
  elif x_test[test_num].split('.')[0] == 'dog' :
    image_data = tf.gfile.FastGFile(test_path + '/'  + image_path[1] + '/' + x_test[test_num], 'rb').read()
    cat_dog_path = 'dogs'
    
  else :
    print('image_data error!!')
    
  with tf.Session() as sess:
    softmax_tensor = sess.graph.get_tensor_by_name('final_result:0')
    predictions = sess.run(softmax_tensor, feed_dict={'DecodeJpeg/contents:0': image_data})
    predictions = np.squeeze(predictions)
  
    argmax = sess.run(tf.argmax(predictions))
  
  # 예측 값과 실제 값 비교
  if int(argmax) == int(y_test[test_num]):
    predict += 1
  
  # 예측이 틀린 경우 그 이미지 호출
  else :
    print('number : ', test_num)
    print('predict label :', argmax)
    
    img = mpimg.imread(test_path + '/' + cat_dog_path + '/' + x_test[test_num])
    imgplot = plt.imshow(img)
    plt.show()
  
  # 30개 마다 호출
  if test_num % 30 == 0 :
    prediction_accuracy = predict / (test_num + 1)
    print('number : ', test_num)
    print('prediction_accuracy : ', prediction_accuracy)
    print('predict label :', argmax)
    
    img = mpimg.imread(test_path + '/' + cat_dog_path + '/' + x_test[test_num])
    imgplot = plt.imshow(img)
    plt.show()

print('learning finish!!')    
```
